#### Replaces blank values with NaN

```python
df=df.replace(r'^\s*$', np.nan, regex=True)
```

#### Removes Rows that contains NaN values

```python
df.dropna(inplace = True)
```

#### Change Age and hrsStudy Columns from object to int

```python
df['Age'] = df['Age'].astype(str).astype(int)
```

```python
df['hrsStudy'] = df['hrsStudy'].astype(str).astype(int)
```

# Task 3

#### Show outliers with a diagram

```python
df["FinalGrade"].plot.box()
```

```python
plt.show()
```

#### Shows which values are NOT outliers

```python
true_index = (1 < df["FinalGrade"].values) &\
             (df["FinalGrade"].values <= 100)
true_index
```

```python
df["FinalGrade"][true_index]
```

#### Shows which values are outliers

```python
false_index = ~true_index
false_index
```

```python
df["FinalGrade"][false_index]
```

#### Shows the index values of the outliers

```python
df["FinalGrade"][false_index].index
```

#### Assign the index values to a variable

```python
outlier_index = df["FinalGrade"][false_index].index
```

#### Removes the outliers from  the data frame

```python
def remove (df, outlier_index):
    outlier_index = sorted(set(outlier_index))
    df = df.drop(outlier_index)
    return df
```

```python
df_cleaned = remove(df, outlier_index)
```

```python
df_cleaned
```

#### Shows diagram after cleaning the data frame

```python
df_cleaned["FinalGrade"].plot.box()
```

```python
plt.show()
```

# Task 4

### Conditions and values for the new column 'Grade'

```python
grade = df_cleaned["FinalGrade"]
conditions = [
    (grade >= 91) & (grade <= 100),
    (grade >= 81) & (grade <= 90),
    (grade >= 71) & (grade <= 80),
    (grade >= 61) & (grade <= 70),
    (grade >= 51) & (grade <= 60),
    (grade <= 50)
]
values = ["A", "B", "C", "D", "E", "F"]
```

```python
df_cleaned["Grade"] = np.select(conditions, values)
```
